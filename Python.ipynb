```python
def palindrome(x):
  res=0
  n=x
  while x>0:
    div=x%10
    res=res*10+div
    x=x//10
  if res==n:
    return "Palindrome"
  else:
    return "Not a Palindrome"



```

```python
x=int(input())
palindrome(x)
```
